# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# explict plot
x = np.linspace(0, 2, 20)

# fig = plt.figure() # empty figure with no axes
fig, ax = plt.subplots() # figure with one axes
ax.plot(x, x**2)
```

```python
# implicit plot, use pyplot to plot, no need to manage figure and axes

plt.plot(x, x**2)
```

### Line plot

```python
fig, ax = plt.subplots(figsize=(6, 3))

ax.set_title('Styling line plot', fontsize=15, color='darkred')

ax.plot(x, x**2, label='y = x^2', color='red', linestyle='--', linewidth=2)
ax.plot(x, x**3, label='y = x^3', marker='v', markersize=5)

ax.set_xlabel('x', fontsize=12, color='darkblue')
ax.set_ylabel('y')

ax.annotate('This is a line', xy=(1.5, 4), xytext=(1, 5),
            arrowprops=dict(facecolor='black', shrink=0.05))

# set the limits of x and y axes
ax.set_xlim([0, 2])
ax.set_ylim([0, 10])

# set the ticks of x axes
ax.set_xticks([0, 0.5, 1, 1.5, 2])

ax.grid(True, color='grey', linestyle=':', linewidth=0.5)

ax.legend(loc='best', fontsize=10)

plt.show()
```

### Scatter plot

```python
np.random.seed(0)
x = np.linspace(0, 2, 20)
y = np.random.rand(20)
colors = np.random.rand(20)
sizes = 1000 * np.random.rand(20)

fig, ax = plt.subplots(figsize=(3, 3))

sct = ax.scatter(x, y, c=colors, s=sizes, marker='*', cmap='coolwarm')

fig.colorbar(mappable=sct)

plt.show()
```

### Histogram

```python
np.random.seed(0)
x = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(3, 3))

ax.hist(x, bins=50, alpha=0.7, histtype='step', edgecolor='green')

plt.show()
```

### Bar chart

```python
# Create some data
labels = ["L1", "L2", "L3", "L4", "L5"]
y1 = [20, 34, 30, 35, 27]
y2 = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # label locations
width = 0.25  # bar width

fig, ax = plt.subplots()
bar1 = ax.bar(x, y1, width, label='Group1')
bar2 = ax.bar(x+width, y2, width, label='Group2')

ax.set_ylabel("Dummy numbers")
ax.set_xticks(x+width/2)
ax.set_xticklabels(labels)
ax.legend()

# Display the figure with the bar chart
plt.show()
```
